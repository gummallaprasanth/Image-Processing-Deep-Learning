Resizing the image using pillow method

```python
from PIL import Image

img = Image.open("/content/tiger.jpg") # storng the image in the img variable

# reducing the dimensions of the image
img_resize= img.resize((200, 200))
```

```python
# save the image in another file

img_resize.save("tiger_resize.jpg")
```

```python
# now we have to convert the image tp array and get the resize image shape

img_resize = mpimg.imread("/content/tiger_resize.jpg")
img_resize.shape
```

```python
# now we can get the image of resized
plt.imshow(img_resize )
```

See the image haa been resized and the shape of the image is reduced compare to the original image

## Converting RGB images to grayscale images using openCV

```python
# importing openCv library

import cv2
```

```python
img= cv2.imread("/content/tiger.jpg")
```

```python
img.shape
```

```python
type(img)
```

```python
# convert the image to grayscale

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# print the shape of the image

img_gray.shape
```

3 has been removed beacuse that is coloe image 3 represent(RGB)

```python
# print the type and numpy array

type(img_gray)
```

```python
print(img_gray)  #it gives the array of the image
```
